```python
# Ejemplo 1: Crear un DataFrame de Spark
data = [("Ana", 23), ("Luis", 31), ("Sofía", 29)]
columns = ["Nombre", "Edad"]
df = spark.createDataFrame(data, columns)
display(df)

# Ejemplo 2: Filtrar filas donde la edad es mayor a 25
df_filtrado = df.filter(df["Edad"] > 25)
display(df_filtrado)

# Ejemplo 3: Agrupar por edad y contar
df_agrupado = df.groupBy("Edad").count()
display(df_agrupado)
```
