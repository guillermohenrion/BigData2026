```python
# Ejemplo 1: Crear un DataFrame de Spark
data = [("Ana", 28), ("Luis", 34), ("Sofía", 22)]
columns = ["Nombre", "Edad"]
df = spark.createDataFrame(data, columns)
display(df)

# Ejemplo 2: Filtrar filas donde la edad es mayor a 25
df_filtrado = df.filter(df["Edad"] > 25)
display(df_filtrado)

# Ejemplo 3: Agregar una columna calculada
from pyspark.sql.functions import lit
df_con_columna = df.withColumn("Ciudad", lit("Madrid"))
display(df_con_columna)

# Ejemplo 4: Agrupar y contar
df_group = df.groupBy("Edad").count()
display(df_group)

# Ejemplo 5: Ejecutar una consulta SQL sobre una tabla temporal
df.createOrReplaceTempView("personas")
result = spark.sql("SELECT Nombre, Edad FROM personas WHERE Edad < 30")
display(result)
```
