```python
# Curso básico de Python

# 1. Variables y tipos de datos
nombre = "Juan"
edad = 25
altura = 1.75
es_estudiante = True

# 2. Operaciones básicas
suma = 5 + 3
resta = 10 - 2
multiplicacion = 4 * 6
division = 8 / 2

# 3. Estructuras de control
if edad > 18:
    print("Eres mayor de edad")
else:
    print("Eres menor de edad")

# 4. Bucles
for i in range(5):
    print("Número:", i)

# 5. Funciones
def saludar(nombre):
    print(f"Hola, {nombre}!")

saludar("Ana")

# 6. Listas
frutas = ["manzana", "banana", "naranja"]
for fruta in frutas:
    print(fruta)

# 7. Diccionarios
persona = {"nombre": "Carlos", "edad": 30}
print(persona["nombre"])

# 8. Clases y objetos
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def presentarse(self):
        print(f"Hola, soy {self.nombre} y tengo {self.edad} años.")

p = Persona("Lucía", 22)
p.presentarse()
```
